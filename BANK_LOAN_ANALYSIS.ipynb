```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**IMPORTING FILES**

```
finance = pd.read_csv('/content/Finance_1.csv')
```

```
finance_2 = pd.read_excel('/content/Finance_2.xlsx')
```

**JOINING FILES**

```
loan = pd.merge(finance, finance_2, how='outer')
loan
```

```
loan.info()
```

**GRADE WISE REVOL_BALANCE**

```
grade_count= loan['grade'].value_counts()
grade_count
```

```
grade_wise_balance = loan.groupby('grade')['revol_bal'].sum().reset_index()

print(grade_wise_balance)
```

```
plt.bar(grade_wise_balance['grade'], grade_wise_balance['revol_bal'])
plt.xlabel('Grade')
plt.ylabel('Revolving Balance')
plt.title('Revolving Balance by Grade')
```

**SUB_GRADE_WISE_REVOL_BALANCE**

```
sub_grade_wise_balance = loan.groupby('sub_grade')['revol_bal'].sum().reset_index()

print(sub_grade_wise_balance)
```

```
plt.figure(figsize=(12,6))
sns.scatterplot(
data=sub_grade_wise_balance,
x="sub_grade",
y="revol_bal")
plt.title("Revolving Balance by Sub-Grade")
plt.xlabel("Loan Sub-Grade")
plt.ylabel("Revolving Balance")
plt.show()
```

**VERIFICATION_STATUS**

```
verification_status_count = loan['verification_status'].value_counts()
verification_status_count
```

```
V_status = loan.groupby('verification_status')['total_pymnt'].sum().reset_index()

print(V_status)
```

```
plt.bar(V_status['verification_status'], V_status['total_pymnt'], )
plt.xlabel('Verification Status')
plt.ylabel('Total Payment')
plt.title('Total Payment by Verification Status')
```

**STATE_WISE_LOAN_STATUS**

```
statewise_loan_status = loan.groupby('addr_state')['loan_status'].value_counts().unstack()
statewise_loan_status
```

```
loan['last_pymnt_d'] = pd.to_datetime(loan['last_pymnt_d'], errors='coerce')
loan['month'] = loan['last_pymnt_d'].dt.month_name()
month_status = pd.crosstab(loan['month'], loan['loan_status'])
plt.figure()
month_status.plot(kind='bar', stacked=True)
plt.xlabel("Month")
plt.ylabel("Number of Loans")
plt.title("Month-wise Loan Status")
plt.show()
```

**HOME_OWNERSHIP V/S LAST_PAYMENT_DATE**

```
sns.boxplot(data=loan, x='home_ownership', y='last_pymnt_d')
plt.xlabel('Home Ownership')
plt.ylabel('Last Payment Date')
plt.title('Distribution of Last Payment Date by Home Ownership')
```
