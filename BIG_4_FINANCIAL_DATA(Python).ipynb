Importing libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
financial = pd.read_csv('/content/audit.csv')
financial.head()

```

dataset information.

```
financial.info()
```

checking null values

```
financial.isnull().sum()
```

checking for duplicate rows

```
financial.duplicated().sum()
```

DATA TRANSFORMATION.

```
financial['risk_percentage'] = (financial['HIGH_RISK_CASES'] / financial['TOTAL_AUDIT_ENGAGEMENTS'])*100
```

```
financial['fraud_cases_percentage'] = (financial['FRAUD_CASES_DETECTED'] / financial['TOTAL_AUDIT_ENGAGEMENTS'])*100
```

```
financial.head()
```

```
financial.describe()
```

# DATA VISUALIZING AND EXPLORATORY DATA ANALYSIS.

SCATTER PLOT CHART(EMPLOYEE_WORKLOAD VS CLIENT_SATISFACTION_SCORE).

```
sns.scatterplot(x=financial['EMPLOYEE_WORKLOAD'], y=financial['CLIENT_SATISFACTION_SCORE'])
plt.show()
```

HISTPLOT CHART

```
numeric_cols = financial.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 4, i)
    sns.histplot(financial[col], bins=20, kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

SCATTER PLOT CHART(EMPLOYEE_WORKLOAD VS AUDIT_EFFECTIVENESS_SCORE

```
sns.scatterplot(x=financial['EMPLOYEE_WORKLOAD'], y=financial['AUDIT_EFFECTIVENESS_SCORE'])
plt.show()
```

SCATTER PLOT CHART(RISK_PERCENTAGE VS TOTAL_AUDIT_ENGAGEMENT)

```
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TOTAL_AUDIT_ENGAGEMENTS', y='risk_percentage', data=financial,)
plt.title('Risk Percentage vs Total Audit Engagements')
plt.xlabel('TOTAL AUDIT ENGAGEMENT')
plt.ylabel('RISK PERCENTAGE')
plt.show()
```

SCATTER PLOT CHART(FRAUD_CASES_PERCENTAGE VS TOTAL_AUDIT_ENGAGEMENT)


```
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TOTAL_AUDIT_ENGAGEMENTS', y='fraud_cases_percentage', data=financial,
                color='royalblue', alpha=0.7)
plt.title('Fraud Cases Percentage vs Total Audit Engagements')
plt.xlabel('Total Audit Engagements')
plt.ylabel('Fraud Cases Percentage')
plt.tight_layout()
plt.show()
```
